```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model_vi_en = AutoModelForSeq2SeqLM.from_pretrained("phatjk/vi_en_bart_translate")
model_en_vi = AutoModelForSeq2SeqLM.from_pretrained("phatjk/en_vi_bart_translate")

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large")
```

```python
import gradio as gr
from transformers import pipeline
from langdetect import detect
import torch



device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
translator_en_vi = pipeline("translation", model=model_en_vi.to(device),tokenizer=tokenizer,device=device)
translator_vi_en = pipeline("translation", model=model_vi_en.to(device),tokenizer=tokenizer,device=device)

def generate(text):
    if (detect(text)=="vi"):
      result = translator_vi_en(text,max_length=128)
    else:
      result = translator_en_vi(text,max_length=128)
    return result[0]["translation_text"]

examples = [
    ["Xin chào, bạn tên là gì?"],
    ["Hôm nay em đến trường"],
    ["Một con chim bay ngang qua đầu tôi"],
    ["The Moon's orbit around Earth has"],
    ["The smooth Borealis basin in the Northern Hemisphere covers 40%"],
    ["Văn minh Maya là một nền văn minh Trung Bộ châu Mỹ của người Maya, nổi bật với hệ chữ tượng hình ký âm—hệ chữ viết tinh vi bậc nhất ở Châu Mỹ thời tiền Columbus—cùng với những thành tựu về nghệ thuật, kiến ​​trúc, toán học, lịch đếm, và chiêm tinh rất phát triển."]
]

demo = gr.Interface(
    fn=generate,
    inputs=gr.inputs.Textbox(lines=5, label="Auto detect (English or Vietnamese)"),
    outputs=gr.outputs.Textbox(label="Translate"),
    examples=examples,
    theme="light"
)

demo.launch(share=True,inbrowser=True)
```
